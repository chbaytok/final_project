```python
import numpy as np
import matplotlib.pyplot as plt
from src.lane_emden import solve_lane_emden
```

```python
# Solve for n = 1
sol = solve_lane_emden(1.0, xi_max=30.0)
print('xi1 =', sol.xi1)
print('mass coeff =', sol.mass_coeff)

plt.figure()
plt.plot(sol.xi, sol.theta, label='numeric')
plt.axhline(0)
plt.xlabel(r'$\xi$')
plt.ylabel(r'$\theta$')
plt.legend()
plt.savefig("../figures/lane_emden_n1_theta.png", dpi=200, bbox_inches="tight")
plt.show()
```

```python
# Analytic comparison for n=1
xi = sol.xi
theta_analytic = np.sin(xi)/xi
theta_analytic[0] = 1.0

plt.figure()
plt.plot(xi, sol.theta, label='numeric')
plt.plot(xi, theta_analytic, '--', label='analytic')
plt.legend()

plt.savefig("../figures/lane_emden_n1_validation.png", dpi=200, bbox_inches="tight")
plt.show()

print('max error =', np.max(np.abs(sol.theta-theta_analytic)))
```
